## 1. Docker Desktop 실행하기

* Docker Desktop 을 실행하면 먼저 Docker ID, Password를 입력해야 한다.
* Docker 에서 자주 사용하는 커맨드 문서로 연결이 가능하다.
![login](fig 03/fig3-1.png)
<center>Login</center>

* Docker를 사용하기 위해서는 충분한 가용 RAM이 있어야 한다.
* RAM이 충분하지 않을 때는 오류 메세지가 뜬다.
![Not enough RAM](fig 03/fig3-2.png)
<center>Not enough RAM</center>

* RAM을 조정하기 위해 먼저 컴퓨터 `작업관리자 - 성능 - 메모리`에서 Docker에 분배할만한 양의 RAM을 계산한다.
![작업관리자](fig 03/fig3-3.png)
<center>작업관리자</center>

* `Docker 아이콘 마우스 우클릭 - Settings - Advanced` 에서 CPU나 Memory 등 Docker 에서 사용하는 자원의 제한을 조절할 수 있다.
* Docker Desktop 실행을 위해 Memory 를 최소치인 1GB로 설정한다.
![Docker icon](fig 03/fig3-4.png)
<center>Docker icon</center>

![Docker settings](fig 03/fig3-5.png)
<center>Docker settings</center>


## 2. Docker Desktop Icon 기능

* 앞서 본 `Settings` 와 같이 Docker Desktop 의 Icon 우클릭으로 실행할 수 있는 기능들에 대해서 살펴본다.
![Docker icon](fig 03/fig3-4.png)
<center>Docker icon</center>

* `About Docker Desktop`
  * 현재 사용하고 있는 Docker Desktop 에 대한 정보를 볼 수 있다.
  * `Release Notes` : Docker Desktop 의 버젼에 따라 수정된 내용을 볼 수 있다.
  * `Acknowledgments` : Docker Desktop 이 취하고 있는 라이센스를 볼 수 있으며 The MIT License 를 따른다.
  * `License Agreement` : 라이센스 및 프로그램 사용에 대한 규약을 볼 수 있다.
![About Docker Desktop](fig 03/fig3-6.png)
<center>About Docker Desktop</center>

* `Discover Docker Enterprise Edition`
  * Enterprise 버젼의 도커를 체험판으로 사용할 수 있다.
![Discover Docker Enterprise Edition](fig 03/fig3-7.png)
<center>Discover Docker Enterprise Edition</center>

* `Settings - General`
  * Docker Desktop 의 기본적인 환경을 설정할 수 있다.
  * 로그인 시 자동 시작 / 자동 업데이트 검사 / 오류 보내기
  * WSL(Windows Subsystem for Linux) 에서 Docker 명령을 실행할 때는 `Expose daemon on tcp://localhost:2375 without TLS` 에 체크해야 한다.
  * Daemon: 멀티태스킹 운영 체제에서 사용자가 직접 제어하지 않고 백그라운드에서 돌면서 여러 작업을 하는 프로그램이다.
  * TLS(Transport Layer Security): 전송 계층 보안은 암호 규약 중 하나이다. 클라이언트/서버 응용 프로그램이 네트워크로 통신을 하는 과정에서 도청, 간섭, 위조를 방지하기 위해서 설계되었다.
![Settings - General](fig 03/fig3-8.png)
<center>Settings - General</center>

* `Settings - Shared Drives`
  * Docker 에서 사용하기 위한 파일들을 복사할 필요 없이 컴퓨터에서 공유하는 용도이다.
  * 공유를 하기 위해서 사용자 이름과 암호를 입력해야 한다.
![Settings - Shared Drives](fig 03/fig3-9.png)
<center>Settings - Shared Drives</center>

* `Settings - Network`
  * Docker 컨테이너가 사용할 네트워크 망에 대해 설정할 수 있다.
![Settings - Network](fig 03/fig3-10.png)
<center>Settings - Network</center>

* `Settings - Proxies`
  * Docker 이미지를 pull 하기 위해 사용할 프록시를 설정할 수 있다.
![Settings - Proxies](fig 03/fig3-11.png)
<center>Settings - Proxies</center>

* `Settings - Daemon`
  * insecure registry: 레지스트리 서버를 구축한 뒤 시스템에 인증서가 필요 없도록 사설 인증서를 등록한다.
  * Docker Hub 에 있는 저장소를 미러링(Clone) 할 수 있다.
![Settings - Daemon](fig 03/fig3-12.png)
<center>Settings - Daemon</center>
  
* `Settings - Kubernetes`
  * Kubernetes: 디플로이 자동화, 스케일링, 컨테이너화된 애플리케이션의 관리를 위한 오픈 소스 시스템이다. 여러 클러스터의 호스트 간에 애플리케이션 컨테이너의 배치, 스케일링, 운영을 자동화 하기 위한 플랫폼을 제공하기 위해 사용한다.
  * `Enable Kubernetes` 를 체크하면 아래 두 체크박스가 활성화 된다.
![Settings - Kubernetes](fig 03/fig3-13.png)
<center>Settings - Kubernetes</center>

* `Settings - Reset`
  * Docker Desktop 을 재시작할 수 있다.
  * 설정사항을 초기화 할 수 있다.
![Settings - Reset](fig 03/fig3-14.png)
<center>Settings - Reset</center>

* `Check for Updates`
  * Docker Desktop 의 버젼을 검사할 수 있다.
![Check for Updates](fig 03/fig3-15.png)
<center>Check for Updates</center>

* `Diagnose and Feedback...`
  * Docker Desktop 프로그램 오류 등을 진단하거나 오류를 제보할 수 있다.
![Diagnose and Feedback](fig 03/fig3-16.png)
<center>Diagnose and Feedback</center>

* `Switch to Windows Containers...`
![Diagnose and Feedback](fig 03/fig3-16.png)
<center>Diagnose and Feedback</center>
