## 1. 도커란?


* 컴퓨터 운영 체제는 일반적으로 가상 메모리를 커널 공간과 사용자 공간으로 나눈다.
![OS](fig 01/fig1-1.png)
<center>운영 체제(OS, Operating System)</center>
* 가상 메모리는 RAM을 관리하는 방법의 하나로, 각 프로그램에 실제 메모리 주소가 아닌 가상의 메모리 주소를 주는 방식이다.
![가상 메모리](fig 01/fig1-2.png)
<center>가상 메모리</center>
* 가상 메모리는 멀티태스킹 운영 체제에서 실제 컴퓨터메모리(주기억장치)보다 큰 메모리 영역을 제공하는 방법으로 사용된다.
* 커널 공간은 커널, 커널 확장 기능, 대부분의 장치 드라이버를 실행하기 위한 예비 공간이다.
* 사용자 공간은 모든 사용자 모드 응용 프로그램들이 동작하는 메모리 영역이다.
![커널 공간, 사용자 공간](fig 01/fig1-3.png)
<center>커널공간과 사용자 공간</center>

* 커널
  * 커널은 컴퓨터의 운영 체제의 핵심이 되는 컴퓨터 프로그램의 하나로, 시스템의 모든 것을 완전히 통제한다.
  * 커널은 컴퓨터 하드웨어와 프로세스의 보안을 책임진다.
  * 한정된 시스템 자원을 효율적으로 관리하여 프로그램의 실행을 원할하게 한다.
  * 프로세스에 처리기를 할당하는 것을 스케줄링이라 한다.
  * 운영 체제의 복잡한 내부를 감추고 깔끔하고 일관성있는 인터페이스를 하드웨어에 제공하기 위해 하드웨어 추상화들로 구현된다.
  * 하드웨어 추상화: 특정 플랫폼의 구체적인 부분과 하드웨어의 자원을 흉내내는 소프트웨어들의 집합이다.
![Kernel](fig 01/fig1-4.png)
<center>커널</center>


* 가상머신(Virtual Machine, VM), 인스턴스
  * 컴퓨팅 환경을 소프트웨어로 구현한 것으로 컴퓨터를 에뮬레이션하는 소프트웨어이다.
  * 에뮬레이터: 한 시스템에서 다른 시스템을 복제한다. 두 번째 시스템이 첫 번째 시스템을 따라 행동하는 것으로 시뮬레이션과는 정 반대이다.
  * 가상머신 상에서 운영체제나 응용 프로그램을 설치 및 실행할 수 있다.
![가상머신](fig 01/fig1-5.png)
<center>가상머신</center>
![가상머신](fig 01/fig1-6.png)
<center>가상머신</center>
![도커와 가상머신](fig 01/fig1-7.png)
<center>도커와 가상머신</center>



* 컨테이너, 운영 체제 수준 가상화(Operating-system-level virtualizetion)
  * 운영 체제의 커널이 하나의 사용자 공간 가상머신을 갖출 수 있도록 하는 서버 가상화 방식이다.
  * 가상 호스팅 환경에 사용하며, 상호 신뢰하지 않는 수많은 사용자들 간에 한정된 하드웨어 자원을 안전하게 할당하는데 유용하다.
  * 시스템 관리에 있어 개별 호스트 상의 서비스들을 특정 서버의 컨테이너로 이동함으로써 서버 하드웨어를 개선하기 위해 이용할 수 있다.
  * 보안 개선, 하드웨어 독립, 자원 관리 기능 추가를 위해 여러 애플리케이션들을 개개의 컨테이너로 분리시킬 수 있다.
  * 즉, 컨테이너는 소프트웨어 구성 요소를 현대적인 엔터프라이즈 환경에 필요한 다양한 애플리케이션과 서비스 스택으로 통합하고, 이런 소프트웨어 구성 요소를 계속 업데이트된 상태로 유지 및 관리할 수 있는 아주 효율적이고 점진적인 메카니즘을 제공한다.
![컨테이너](fig 01/fig1-8.png)
<center>Container</center>

  
* 도커는 리눅스의 응용 프로그램들을 소프트웨어 컨테이너 안에 배치시키는 일을 자동화하는 오픈 소스 프로젝트이다.
* 도커 컨테이너는 소프트웨어를 소프트웨어의 실행에 필요한 모든 것을 포함하는 완전한 파일 시스템 안에 감싼다. 코드, 런타임, 시스템 도구, 시스템 라이브러리 등 서버에 설치되는 무엇이든 포함한다.
* 이러한 컨테이너를 통해 실행 중인 환경에 관계 없이 언제나 동일하게 실행될 것을 보증한다.
* 도커는 리눅스에서 컨테이너의 추상화 및 자동화 계층을 추가적으로 제공한다.
* 도커를 통해 애플리케이션을 신속하게 구축, 테스트 및 배포할 수 있다.
* 도커를 사용하면 환경에 구애받지 않고 애플리케이션을 신속하게 배포 및 확장할 수 있으며 코드에 이상없이 실행할 수 있다.
* 도커는 코드를 실행하는 표준 방식을 제공하며, 이는 곧 컨테이너를 위한 운영체제이다.
* 도커는 각 서버에 설치되며 컨테이너를 구축, 시작, 중단 하는 데 사용할 수 있는 간단한 명령을 제공한다.
![docker](fig 01/fig1-9.png)
<center>Docker</center>


* 도커는 애플리케이션 뿐만 아니라 실행에 필요한 시스템 환경을 모아서 컨테이너(Container)로 관리한다.
* 이렇게 만든 것을 도커 이미지(Docker Image)라고 하는데 이 이미지로 만든 컨테이너는 도커가 설치된 곳이라면 어디든 똑같이 동작한다.
![docker](fig 01/fig1-10.png)
<center>Docker</center>


## 2. 간단히 살펴보는 도커


* 절차 단순화
  * 작업 과정과 커뮤니케이션을 모두 단순화할 수 있어 배포하기 쉽다.
![전통적인 워크플로](fig 01/fig1-11.png)
<center>전통적인 워크플로</center>
![도커 워크플로](fig 01/fig1-12.png)
<center>도커 워크플로</center>
  * 헤로쿠 같은 푸시-투-디플로이 시스템은 개발자에게 애플리케이션, 의존성을 제어할 수 있도록 해준다. 도커는 이러한 헤로쿠와 유사한 방식으로 생산성을 향상시키도록 역할 분리와 의존성의 캡슐화를 제공한다. git push 만으로 배포가 해결되는 시스템이다.
  * 도커는 '탈부착 가능한 배터리 포함'의 접근 방식을 고수함으로써 사용자가 여러 환경에 필요한 요구 사항들을 만족시킬 수 있도록 한다.
  * 모든 의존성 이슈를 개발 및 테스트 단계에서 발견하는 것이 가능하다.


* 폭 넓은 지원과 채택
  * AWS 일래스틱 빈스토크, 구글 앱 엔진, IBM 클라우드, 마이크로소프트 애저, 랙스페이스 클라우드 등 유명한 대형 클라우드들이 도커에 대한 지원을 발표하였다.
  * 도커의 컨테이너는 클라우드 공급자들에게 '한 번 작성해서 모든 곳에서 실행 가능한' 클라우드 애플리케이션을 만들 수 있는 표준 포맷이 되어가는 중이다.
  * 도커는 리눅스 시스템을 기반으로 만들어졌지만 현재 다양한 운영체제에서 사용이 가능하다.


* 아키텍처
  * 도커는 기술이 강력한 만큼 어느정도 진입 장벽을 가지고 있다.
  * 도커 기반의 아키텍처는 도커 커맨드의 실행 방식에 따라 서버와 클라이언트 모두 동작하는 하나의 실행 파일로 이루어진 단순한 클라이언트/서버 모델이다.


* 도커 활용하기
  * 컨테이너 생성은 매우 작은 공간을 필요로 하여 꽤 가볍다.
  * 새로운 컨테이너는 계층화된 파일 시스템, 설정 데이터 정도만 있으므로 생성 및 삭제가 용이하다.
  * 다양한 환경에 거리낌 없이 컨테이너 내에 모든 애플리케이션을 배치하고 불변의 인프라를 구성하여 단순화할 수 있다.
