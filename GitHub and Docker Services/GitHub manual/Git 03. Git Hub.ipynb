## 1. 가입하기

1. GitHub 사이트에 접속한다. (https://github.com)
2. `Sign up for GitHub`를 클릭한다.
![github.com](fig 03/fig3-1.png)


3. 개인정보(이름, 이메일, 비밀번호, 매크로 확인)를 입력한다.
 * 이름의 경우 빈칸을 사용하면 안되며 다른 사용자와 중복 불가능하다.
 * GitHub 저장소의 링크가 `github.com/<사용자명>/<프로젝트명>` 으로 이루어지기 때문에 이름 설정을 할 때 유의 하여야 한다.
 * 입력한 이메일로 인증 메일이 수신되므로 정확한 이메일을 기재하여야 한다.
 * 매크로 확인 과정에서 귀여운 말을 똑바로 세우도록 돌려주면 된다.
 * 일부 인터넷 프로그램은 매크로 확인 부분이 제대로 보여지지 않으므로 그럴 때는 다른 인터넷을 사용하도록 한다.(Internet Explorer 대신 Chrome 등을 권장)
![sign up](fig 03/fig3-2.png)


4. Free/Pro를 선택한다.
 * 무료 계정으로도 GItHub 대부분의 기능을 사용할 수 있다.
 * 무료 계정은 모든 사람이 읽을 수 있는 공개 프로젝트 혹은 자신만 읽을 수 있는 Private 프로젝트만 만들 수 있다.
 * GitHub에 돈을 내면 비공개 프로젝트(협업)도 만들 수 있다.
![Free/Pro](fig 03/fig3-3.png)


5. 간단한 설문에 참여한다.
 * 건너뛰기도 가능하다.
![question](fig 03/fig3-4.png)


6. 이메일로 인증한다.
![verify](fig 03/fig3-5.png)




## 2. 프로젝트에 기여하기


* GitHub 플로우 (프로젝트에 기여하는 방법)
 1. 프로젝트를 `Fork`한다.
 2. `master` 기반으로 토픽 브랜치 만든다.
   * 토픽 브랜치: 프로젝트 원본인 마스터 브랜치에서 특정 주제의 부분만을 새로 개선하여 프로젝트를 수정시키기 위한 발판이다.
 3. 토픽에 맞도록 수정하여 커밋한다.
 4. 자신의 GitHub 프로젝트에 브랜치를 `Push`한다.
 5. GitHub에 `Pull Request`생성한다.
 6. 사용자와 프로젝트 소유자가 토론하면서 그에 따라 계속 커밋하며 반복한다.
 7. 프로젝트 소유자는 `Pull Request`를 `Merge`한다.
![github flow1](fig 03/fig3-6.png)
![github flow2](fig 03/fig3-7.png)



* 프로젝트에 Fork 하기
 * 참여하고 싶은 프로젝트가 생기면 "Fork" 한다.
 * Fork하면 GitHub이 프로젝트를 통째로 복사해준다.
 * 복사본은 특정 폴더를 점유하여 네임스페이스를 가진다.
 * 네임스페이스: 개체(여기서는 프로젝트 복사본)를 구분할 수 있는 단위이며 하나의 이름이 단 하나의 개체만을 가리킨다.
 * 프로젝트 페이지를 방문해서 오른쪽 꼭대기에 있는 `Fork`버튼을 클릭하면 된다.
![fork](fig 03/fig3-8.png)
![forking](fig 03/fig3-9.png)
![after fork](fig 03/fig3-10.png)
![comparing fork](fig 03/fig3-11.png)
 * Fork를 하여 생긴 복사본은 사용자의 저장소에 저장된다. (즉 사용자의 소유이고, `Settings` 기능을 쓸 수 있다.)
 * 어디서 Fork를 해온건지 알 수 있다.
 * 자신의 프로젝트는 Fork할 수 없다.
 * 기존의 저장소와 이름이 같은 저장소는 `-1`이 붙은 상태로 복사된다.
 
 
* Fork 방식의 이해
 * 프로젝트 소유자는 다른 사용자들을 프로젝트에 추가하거나 Push 권한을 줄 필요가 없다.
 * 사용자는 자발적으로 프로젝트를 Fork해서 Push할 수 있다.
 * Push한 변경 내용을 원래 저장소로 보내 기여한다.
 * 이러한 방식을 Pull Request라고 한다.
 * 토론 스레드를 만들고 거기서 리뷰를 하면서 토론할 수 있다.
 * 프로젝트 소유자 맘에 든다면 Merge 하여 프로젝트 원본을 수정할 수 있다.


* 토픽 브랜치 만들고 Publish 하기
 * GitHub Desktop에서 `Branch`-`New branch` (`Ctrl`+`Shift`+`N`)
![new branch](fig 03/fig3-12.png)
 * 토픽 브랜치의 이름을 입력하고 `Create branch`를 클릭한다.
 * `Publish branch`를 누르면 GitHub 저장소에 새로운 브랜치가 개설된다. (`Ctrl`+`P`)
![publish branch](fig 03/fig3-13.png)


* 수정 후 Commit 하기
 * 파일을 수정하거나 추가하면 Commit 기능이 활성화 된다.
 * 커밋할 준비가 된 상태는 Staged 이다.
![commit](fig 03/fig3-14.png)


* GItHub로 Push 하기
 * 커밋 후 수정된 프로젝트 브랜치를 다시 GItHub로 Push해야 한다.
 * `Push origin`을 누르면 브랜치의 내용이 커밋된 내용으로 수정된다. (`Ctrl`+`P`)
![push](fig 03/fig3-15.png)


* Pull Request 만들기
 * Push 하고 나면 GitHub 사이트에서는 `Compare & Pull Request` 기능이 활성화 된다.
 ![compare & pull request](fig 03/fig3-16.png)
 * 혹은 `branches`를 누르면 사용자의 브랜치에 대한 상태를 볼 수 있다.
 ![branches](fig 03/fig3-17.png)
 * `Compare & Pull Request` 혹은 `New pull request`를 누르면 프로젝트의 어떤 브랜치를 선택할 지와 Pull request의 제목 및 커멘트, 수정사항 등을 보여준다.
 ![open pull request](fig 03/fig3-18.png)
 * Pull request를 완료하면 프로젝트 소유자와 토론할 수 있으며 Merge 혹은 거절 등의 대답을 기다린다.
 ![complete pull request](fig 03/fig3-19.png)


* Pull Request 처리하기
 * 누군가가 Pull request 하고 나면 `Pull requests` 뒤의 수가 달라진다.
 ![pull requests](fig 03/fig3-20.png)
 * `Pull requests`를 클릭하면 구체적인 내역을 볼 수 있다.
 ![check pull requests](fig 03/fig3-21.png)
 * 누군가가 올린 Pull request를 클릭하면 Merge하거나 코멘트를 달 수 있다.
 ![merge & comment](fig 03/fig3-22.png)
