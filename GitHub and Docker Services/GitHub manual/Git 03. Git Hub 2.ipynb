## 2. 로컬 저장소 만들기


1. 로컬 디렉토리를 Git 저장소로 만들기(Publish)
2. 온라인에 있는 Git 저장소를 복사 하기(Clone)
![저장소 만들기](fig 03/fig3-9.png)
<center>로컬 저장소를 만드는 원리</center>

Publish/Clone 두 가지 원리로 GitHub Desktop을 통해 Git 저장소를 만들 수 있다.

![GitHub Desktop으로 저장소 만들기](fig 03/fig3-10.png)
<center>GitHub Desktop을 처음 실행했을 때(로컬 저장소가 없는 상태)</center>


 * `Your repositories`
  * GitHub 계정에 등록되어 있는 저장소를 Clone 할 수 있다.
  * 단축키 `Ctrl`+`Shift`+`O` 혹은 `Files`-`Clone repository...`로도 가능하다.
![나의 저장소](fig 03/fig3-11.png)
<center>내 저장소에서 Clone 하기</center>


 * `Clone a repository from the Internet...`
  * 인터넷 상의 저장소에서 받아온 링크를 통해 해당 저장소를 Clone 할 수 있다.
  * `URL`에서는 GitHub 사이트 상의 링크(사용자명/프로젝트명)을 입력하면 손쉽게 Clone할 수 있다.
  * `GitHub.com`에서는 GitHub Desktop을 로그인한 계정의 저장소를 선택하여 불러올 수 있으며 위에 있는 `Your repositories`와 비슷하다.
![리모트 저장소 Clone](fig 03/fig3-12.png)
<center>인터넷 상의 저장소 Clone 하기</center>


 * `Create a New Repository on your hard drive...`
  * 새로운 저장소를 로컬 디렉토리 상에 만들 수 있다. 
  * 만든 후에 `Publish`를 하면 인터넷 상에 저장소가 등록된다.
  * 단축키 `Ctrl`+`N` 혹은 `Files`-`New repository...`로도 가능하다.
![새로운 로컬 저장소](fig 03/fig3-13.png)
<center>로컬 저장소를 새로 만들기</center>


 * `Add an Existing Repository from your hard drive...`
  * 로컬 디렉토리 상에 이미 존재하는 폴더를 저장소로 만들 수 있다.
  * 만든 후에 `Publish`를 하면 인터넷 상에 저장소가 등록된다.
  * 단축키 `Ctrl`+`O` 혹은 `Files`-`Add local repository...`로도 가능하다.
![기존의 로컬 저장소](fig 03/fig3-14.png)
<center>기존 로컬 디렉토리를 저장소로 등록하기</center>

## 3. 수정하고 저장소에 저장하기

파일을 수정하고 파일의 스냅샷을 커밋하기

* 파일의 세 가지 상태 (복습)
 * Committed: 데이터가 로컬 데이터베이스에 안전하게 저장됨
 * Modified: 수정한 파일을 아직 커밋하지 않은 상태
 * Staged: 현재 수정한 파일을 곧 커밋할 것이라고 표시한 상태
<br>
![파일의 세 가지 상태](fig 01/fig1-6.png)
<center>워킹 트리, Staging Area, Git 디렉토리</center>
<br>

* 워킹 디렉토리의 파일은 Tracked(관리대상임) Untracked(관리대상이 아님) 으로 나눔
* Tracked 파일은 이미 스냅샷에 포함되어 있던 파일이며 Unmodified(수정하지 않음)와 Modified(수정함) 그리고 Staged(커밋으로 저장소에 기록할 상태) 으로 나눔
* Untracked 파일은 워킹 디렉토리에 있는 파일 중 스냅샷, Staging Area 어디에도 포함되지 않은 파일임
* 처음 저장소를 Clone 하면 모든 파일은 Tracked - Unmodified 상태(Checkout 이후 수정하지 않았으므로)
* 커밋 이후 파일을 처음으로 수정하면 Git은 그 파일을 Modified로 인식함
* 커밋을 하기 위해서는 이러한 Modified 파일을 Staged 상태로 만든 뒤에 Staged 상태인 파일을 커밋함

![파일의 라이프사이클](fig 03/fig3-19.png)
<center>파일의 라이프사이클</center>


* 파일의 상태 확인하고 커밋하기

 * 파일을 수정하거나 추가하면 Commit 기능이 활성화 된다.
 * 커밋할 준비가 된 상태는 Staged 이다.
 * 커밋할 때 커밋에 대한 제목(필수)과 내용(옵션)을 입력한다.
 * `+` 저장소에 새로운 파일이 추가 되었음을 의미한다.
 * `-` 저장소에 있던 파일이 삭제 되었음을 의미한다.
 * `●` 저장소에 있던 파일이 수정 되었음을 의미한다.
![commit](fig 02/fig2-14.png)
<center>파일의 상태 확인하고 커밋하기</center>


## 4. 커밋 히스토리 조회하기

* 저장소에서 `commits`를 누르면 커밋한 모든 시점의 기록을 볼 수 있다.
![commits1](fig 03/fig3-20.png)
<center>Git 저장소 메인에서 commits 누르기</center>


![commits2](fig 03/fig3-21.png)
<center>commits를 누른 뒤 화면</center>


 * `Verified` 믿을만한 사용자가 커밋했음을 알려주는 증표이다.
 * GPG key를 통해 커밋한 사용자에 대한 정보를 얻을 수 있다.
 * https://help.github.com/articles/managing-commit-signature-verification/ 에서 자세한 설명을 볼 수 있다.


* `<>`(Browse files) 아이콘을 누르면 커밋한 시점의 저장소 스냅샷을 보여준다.
 * 해당 시점의 저장소를 Clone 할 수 있다.
![commits3](fig 03/fig3-22.png)
<center>Browse files로 특정 시점의 파일 확인하기</center>


* 서류 아이콘을 누르면 해당 커밋에 대한 40자의 체크섬(1장 3절 참조)을 복사해준다.
 * 아이콘을 누른 결과 `c9ba2cb100af31f24496eb6bfe927cd90e45e1b9`가 자동 복사 되었다.


* `c9ba2cb`와 같은 체크섬(실제로 40자이지만 앞 7자를 따서 보여줌)을 누르면 어떤 내용을 커밋하였는지 보여준다.
 * `Browse files`를 누르면 커밋한 시점의 저장소 스냅샷을 보여준다.
 * 해당 시점의 저장소를 Clone 할 수 있다.
![commits4](fig 03/fig3-23.png)
<center>커밋 내용 확인하기</center>


* 커밋에 대한 코멘트를 보거나 작성할 수 있다.
 * `Lock conversation` 다른 사용자가 새로운 코멘트를 작성할 수 없도록 하며, 프로젝트 소유자 또는 협업자는 여전히 코멘트를 작성할 수 있다. Unlock도 가능하다.


* `Subscribe` 를 누르면 해당 커밋에 대한 알림을 받을 수 있다.



## 5. 되돌리기

* 커밋을 하고 `Push` 하지 않은 경우에는 GitHub Desktop 에서 Commit 도구 창 아래에 `UNDO`를 누르면 커밋하기 이전으로 되돌릴 수 있다.
![UNDO](fig 03/fig3-24.png)
<center>Push 하기 이전의 되돌리기</center>


* 커밋을 하고 `Push` 한 이후 되돌리기
  1. github.com 의 저장소에서 `commits`-되돌리고 싶은 시점의 `<>`-`Download ZIP`
    * `commits`의 수가 현재보다 적고, `Tree:c443b7dfa6` 을 보아 현재가 아닌 특정 시점의 저장소 스냅샷을 보여주는 것을 알 수 있다.
![Browse files](fig 03/fig3-25.png)
<center>Browse files 누르기</center><br>
![Download ZIP](fig 03/fig3-26.png)
<center>Download ZIP 누르기</center><br>
  2. 다운로드 받은 ZIP 파일에서 되돌리고 싶은 파일만 로컬 저장소에 덮어쓰기
![backup](fig 03/fig3-27.png)
<center>로컬 저장소에서 파일 덮어쓰기</center><br>
  3. `Commit` 하고 `Push` 하기
    * 오른쪽 상단에 경고 아이콘에 마우스를 가져다 대면 오류의 원인을 알 수 있다.
    * `line endings have changed from lf to crlf` Windows와 Unix, Mac OS의 Line ending 방식에 차이가 있어서 코드의 변경이 없음에도 불구하고 commit 하게 될 수 있어서 변경 로그를 보거나 merge 할 때 마다 문제의 소지가 있음을 경고한다.
    * Commit 불가능한 오류가 아니라 그대로 진행하였다.
![commit&push](fig 03/fig3-28.png)
<center>로컬 저장소에서 파일 덮어쓰기</center>


## 6. 결론

1. Git 저장소 만들기
 * github.com 상에서 저장소를 만들고 설정할 수 있다.
2. 로컬 저장소 만들기
 * GitHub 상의 저장소를 Clone 하여 로컬 환경에 동일한 저장소를 만들 수 있다.
 * 로컬 디렉토리 상의 저장소를 GitHub 상에 등록할 수 있다.
3. 수정하고 저장소에 저장하기
 * GitHub Desktop 으로 수정된 내용을 한 눈에 확인하고 커밋할 수 있다.
4. 커밋 히스토리 조회하기
 * 커밋 내용을 조회하여 특정 시점의 스냅샷을 확인할 수 있다.
5. 되돌리기
 * 잘못된 커밋을 바로잡기 위해 이전 시점의 스냅샷을 불러와서 덮어쓸 수 있다.
