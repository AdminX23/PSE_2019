## 2. GitHub의 워크플로우


* GitHub 플로우 (프로젝트에 기여하는 방법)
 1. 프로젝트를 `Fork`한다.
 2. `master` 기반으로 토픽 브랜치 만든다.
   * 토픽 브랜치: 프로젝트 원본인 마스터 브랜치에서 특정 주제의 부분만을 새로 개선하여 프로젝트를 수정시키기 위한 발판이다.
 3. 토픽에 맞도록 수정하여 커밋한다.
 4. 자신의 GitHub 프로젝트에 브랜치를 `Push`한다.
 5. GitHub에 `Pull Request`생성한다.
 6. 사용자와 프로젝트 소유자가 토론하면서 그에 따라 계속 커밋하며 반복한다.
 7. 프로젝트 소유자는 `Pull Request`를 `Merge`한다.
![github flow1](fig 02/fig2-6.png)
<center>프로젝트 소유자와 사용자 한 명 사이의 플로우</center>
![github flow2](fig 02/fig2-7.png)
<center>프로젝트 소유자와 사용자 두 명 사이의 플로우</center>

 
* Fork 방식의 이해
 * 프로젝트 소유자는 다른 사용자들을 프로젝트에 추가하거나 Push 권한을 줄 필요가 없다.
 * 사용자는 자발적으로 프로젝트를 Fork해서 Push할 수 있다.
 * Push한 변경 내용을 원래 저장소로 보내 기여한다.
 * 이러한 방식을 Pull Request라고 한다.
 * 토론 스레드를 만들고 거기서 리뷰를 하면서 토론할 수 있다.
 * 프로젝트 소유자 맘에 든다면 Merge 하여 프로젝트 원본을 수정할 수 있다.




## 3. 프로젝트에 기여하기


* 프로젝트에 Fork 하기
 * 참여하고 싶은 프로젝트가 생기면 "Fork" 한다.
 * Fork하면 GitHub이 프로젝트를 통째로 복사해준다.
 * 복사본은 특정 폴더를 점유하여 네임스페이스를 가진다.
 * 네임스페이스: 개체(여기서는 프로젝트 복사본)를 구분할 수 있는 단위이며 하나의 이름이 단 하나의 개체만을 가리킨다.
 * 프로젝트 페이지를 방문해서 오른쪽 꼭대기에 있는 `Fork`버튼을 클릭하면 된다.
![fork](fig 02/fig2-8.png)
<center>저장소에서 Fork 버튼 찾기</center>

![forking](fig 02/fig2-9.png)
<center>내 저장소로 Fork 하는 중</center>

![after fork](fig 02/fig2-10.png)
<center>Fork 완료</center>

![comparing fork](fig 02/fig2-11.png)
<center>내 저장소 관리</center>


 * Fork를 하여 생긴 복사본은 사용자의 저장소에 저장된다. (즉 사용자의 소유이고, `Settings` 기능을 쓸 수 있다.)
 * 어디서 Fork를 해온건지 알 수 있다.
 * 자신의 프로젝트는 Fork할 수 없다.
 * 기존의 저장소와 이름이 중복되는 저장소는 `-1`이 붙은 상태로 복사된다.
 

* 토픽 브랜치 만들고 Publish 하기
 * GitHub Desktop에서 `Branch`-`New branch` (`Ctrl`+`Shift`+`N`)
![new branch](fig 02/fig2-12.png)
<center>새 브랜치 만들고 이름 설정하기</center>


 * 토픽 브랜치의 이름을 입력하고 `Create branch`를 클릭한다.
 * `Publish branch`를 누르면 GitHub 저장소에 새로운 브랜치가 개설된다. (`Ctrl`+`P`)
![publish branch](fig 02/fig2-13.png)
<center>새 브랜치를 Git 서버로 Publish 하기</center>


* 수정 후 Commit 하기
 * 파일을 수정하거나 추가, 삭제하면 Commit 기능이 활성화 된며 변화 내용을 볼 수 있다.
 * 커밋할 준비가 된 상태는 Staged 이다.
 * `+`: 저장소에 새로운 파일이 추가 되었음을 의미한다.
 * `-`: 저장소에 있던 파일이 삭제 되었음을 의미한다.
 * `●`: 저장소에 있던 파일이 수정 되었음을 의미한다.
![commit](fig 02/fig2-14.png)
<center>저장소의 변화를 수용하는 GitHub Desktop</center>


* GItHub로 Push 하기
 * 커밋 후 수정된 프로젝트 브랜치를 다시 GItHub로 Push해야 한다.
 * `Push origin`을 누르면 브랜치의 내용이 커밋된 내용으로 수정된다. (`Ctrl`+`P`)
![push](fig 02/fig2-15.png)
<center>Push 활성화</center>


* Pull Request 만들기
 * Push 하고 나면 GitHub 사이트에서는 `Compare & Pull Request` 기능이 활성화 된다.
 ![compare & pull request](fig 02/fig2-16.png)
<center>Pull Request 활성화</center> 
 
 
 * 혹은 `branches`를 누르면 사용자의 브랜치에 대한 상태를 볼 수 있다.
 ![branches](fig 02/fig2-17.png)
<center>branches에서 브랜치 확인하고 Pull Request 하기</center>
 
 
 * `Compare & Pull Request` 혹은 `New pull request`를 누르면 어떤 프로젝트/브랜치를 선택할 지와 Pull request의 제목 및 커멘트, 수정사항 등을 보여준다.
 ![open pull request](fig 02/fig2-18.png)
<center>Pull Request 하기</center>

 
 * Pull request를 완료하면 프로젝트 소유자와 토론할 수 있으며 Merge 혹은 거절 등의 대답을 기다린다.
 ![complete pull request](fig 02/fig2-19.png)
<center>Pull Request 완료하기</center>

## 5. 결론

 1. 가입하기
  * `github.com`에서 어렵지 않은 방법으로 가입하여 GitHub 대부분의 기능을 이용할 수 있다.
 2. GitHub의 워크플로우
  * GItHub에서 프로젝트 소유자와 사용자의 협업을 쉽게 할 수 있다.
 3. 프로젝트에 기여하기
  * `Fork`한 저장소의 내용을 `Commit`하여 `Pull Request`하여 프로젝트에 기여할 수 있다.
 4. Pull Request 받아들이기
  * 사용자가 요청한 `Pull Request`를 원본 프로젝트와 `Merge`하거나 거절할 수 있다.
